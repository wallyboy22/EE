### 4: Plotting already classified data from the country


```python
# last_update: '2024/10/30', github:'mapbiomas/brazil-fire', source: 'IPAM', contact: 'contato@mapbiomas.org'
# MapBiomas Fire Classification Algorithms Step A_4_1_plot_fire_regions.py
### Step A_4_0 - Functions for Plotting Fire Regions

import geemap
import ee
from datetime import datetime

def plot_fire_regions(country, ee_project):
    """
    Plots the fire region grid for a specific country in Google Earth Engine.

    Parameters:
    - country: Country code or name (string).
    - ee_project: ID of the Google Earth Engine project (string).
    """
    # Create an interactive map
    Map = geemap.Map(center=[0, 0], zoom=2)
    Map.setOptions('SATELLITE')  # Set the base map type (optional)

    # Define the path to the fire regions
    regiones_path = f'projects/mapbiomas-{country}/assets/FIRE/AUXILIARY_DATA/regiones_fuego_{country}_v1'

    try:
        regiones_fc = ee.FeatureCollection(regiones_path)
        print(f"[INFO] Loaded FeatureCollection of regions: {regiones_path}")
    except Exception as e:
        print(f"[ERROR] Failed to load regions: {e}")
        return

    # Add the regions layer to the map with a base color
    Map.addLayer(
        ee.Image().paint(regiones_fc, 'id_region'),
        {'palette': 'FF0000'},  # Define a base color for the regions
        'Fire Regions'
    )

    # Collection of burned area images
    fire_path = f'projects/mapbiomas-{country}/assets/FIRE/COLLECTION1/CLASSIFICATION/burned_area_{country}_v1'

    try:
        fire = ee.ImageCollection(fire_path)
        print(f"[INFO] Loaded ImageCollection of burned areas: {fire_path}")
    except Exception as e:
        print(f"[ERROR] Failed to load burned area collection: {e}")
        return

    # Print distinct regions present in the collection
    try:
        distinct_regions = fire.aggregate_array('region').distinct().getInfo()
        print(f"[INFO] Distinct regions in the fire collection: {distinct_regions}")
    except Exception as e:
        print(f"[ERROR] Failed to retrieve distinct regions: {e}")
        distinct_regions = []

    # Define a color palette for fire layers
    colors = [
        'red', 'blue', 'green', 'yellow', 'purple', 'orange',
        'cyan', 'magenta', 'lime', 'pink', 'teal', 'lavender',
        'brown', 'beige', 'maroon'
    ]

    if not distinct_regions:
        print("[WARNING] No regions found for plotting.")
        return

    # Add the fire layer for all regions using the first color in the palette
    Map.addLayer(
        fire,
        {'palette': [colors[0]]},
        'Fire - All Regions'
    )

    # Add layers filtered by region with colors from the palette
    for i, region in enumerate(distinct_regions):
        color = colors[i % len(colors)]  # Cycle through colors if there are more regions than colors
        region_filter = ee.Filter.eq('region', region)
        filtered_fire = fire.filter(region_filter)

        # Add the filtered fire layer by region
        Map.addLayer(
            filtered_fire,
            {'palette': [color]},
            f'Fire - Region {region} ({color})'
        )

        # Define the path for the specific region's grid
        grid_path = f'projects/mapbiomas-{country}/assets/FIRE/AUXILIARY_DATA/GRID_REGIONS/grid-{country}-{region}'

        try:
            grid_fc = ee.FeatureCollection(grid_path)
            print(f"[INFO] Loaded FeatureCollection grid: {grid_path}")
        except Exception as e:
            print(f"[ERROR] Failed to load grid for region {region}: {e}")
            continue  # Skip to the next iteration if it fails

        # Add the grid layer to the map
        Map.addLayer(
            ee.Image().paint(grid_fc, 'a', 1),
            {'palette': [color]},
            f'Grid - Region {region}'
        )

        # Optionally: Add the grid FeatureCollection as a vector layer
        Map.addLayer(
            grid_fc,
            {'color': color},
            f'Regions - Grid {region}',
            False  # Set to not visible by default
        )

    # Display the map
    return Map

# Example usage of the function
if __name__ == "__main__":
    # Define the necessary variables
    country = 'GUYANA'  # Replace with the correct country code or name
    ee_project = 'your-gee-project-id'  # Replace with your GEE project ID

    # Call the function to plot fire regions
    mapa = plot_fire_regions(country, ee_project)

    if mapa:
        mapa  # Displays the interactive map in Jupyter Notebook or compatible environment
```
